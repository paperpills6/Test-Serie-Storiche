```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
csv_path = r"C:\Users\admloc\Desktop\Python\Test serie storiche\HistoricalData_1760090934890.csv"
df = pd.read_csv(csv_path)
df.head()
```

```python
# Statistical summary of Apple stock prices
summary = df.describe()
print(summary)

# Convert price columns to numeric (remove $ and convert to float)
for col in ['Close/Last', 'Open', 'High', 'Low']:
    df[col] = df[col].replace({'\$': ''}, regex=True).astype(float)

# Plot closing price over time (if column exists)
if 'Close/Last' in df.columns:
    plt.figure(figsize=(12,6))
    plt.plot(df['Date'], df['Close/Last'])
    plt.title('Apple Closing Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.xticks(rotation=45)
    plt.show()
else:
    print('Column "Close/Last" not found in the data.')

# Correlation heatmap (only numeric columns)
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
